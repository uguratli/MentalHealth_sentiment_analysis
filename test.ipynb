```python
import nltk
nltk.download('punkt_tab')
```

```python
def sentence_count(x):
    return nltk.sent_tokenize(x)
```

```python
df["char_count"] = df["statement"].str.len()
df["sentences"] = df["statement"].apply(sentence_count)
df["sentence_count"] = df["sentences"].str.len()
df.head()
```
